```python
def makepoints(x, y):
    return (x-5)**2+(y-5)**2

fig = plt.figure(figsize=(10,4))
ax1 = plt.subplot(121, projection='3d')
ax2 = plt.subplot(122)

p = Point3D(2, 2, 2)
# using Plane()
origin = Point3D(0, 9, 0)
p1 = Plane(origin, normal_vector =(-1, 1, 0))
e1 = [1/np.sqrt(2), 1/np.sqrt(2), 0]
e2 = [0,0,1]
e = np.vstack([e1,e2])


for x in np.arange(0,10):
    y = x
    z = makepoints(x,y)
    ax1.scatter(x, y, z, color="black")
    p3 = p1.projection(Point3D(x, y, z))
    ax1.scatter(p3[0], p3[1], p3[2], color="red")
    # new 2d x coordinate
    #x2d = np.sqrt(float((x-origin[0])**2 + (y-origin[1])**2))
    x2d, z2d = np.dot(np.vstack([e1,e2]),p3)
    plt.plot(x2d,z2d, "ro")
    
    print(p3)

ax1.set_xlabel('X Label')
ax1.set_ylabel('Y Label')
ax1.set_zlabel('Z Label')

plt.tight_layout()

#plt.show()
```

```python
origin = [0,9,0]
#normal = [-1/np.sqrt(2),1/np.sqrt(2),0]
normal = [0,1,0]
e1 = [1/np.sqrt(2), 1/np.sqrt(2), 0]
e2 = [0,0,1]

e1 = [1, 0, 0]
e2 = [0,0,1]


a, b, c = normal[0], normal[1], normal[2]
p, q, r = origin[0], origin[1], origin[2]
d = a*p + b*q + c*r

def project(x, y, z):
    k = (d - a*x- b*y - c*z) / (a**2 + b**2 + c**2)
    return x+k*a, y+k*b, z+k*c


e = np.vstack([e1,e2])

fig = plt.figure(figsize=(10,4))
ax1 = plt.subplot(121, projection='3d')
ax2 = plt.subplot(122)

for x in np.arange(0,10):
    y = x
    z = makepoints(x,y)
    ax1.scatter(x, y, z, color="black")
    p3 = project(x, y, z) 
    ax1.scatter(p3[0], p3[1], p3[2], color="red")
    # new 2d x coordinate
    #x2d = np.sqrt(float((x-origin[0])**2 + (y-origin[1])**2))
    x2d, z2d = np.dot(e,p3)
    plt.plot(x2d,z2d, "ro")

```
